```python
import pandas as pd
import numpy as np

music_df = pd.read_csv('../../data/processed_data/DimLoadingsUSA.csv')
# 빈 컬럼명 추가
music_df.rename(columns={'Unnamed: 0': 'Music_Id'}, inplace=True)

# mood 컬럼 추가
music_mood = music_df.iloc[:,1:]
music_df['mood'] = music_mood.idxmax(axis=1)

# id와 mood제외한 컬럼 삭제
music_df = music_df[['Music_Id', 'mood']]
music_df
# music_df.to_csv('../../data/processed_data/DimLoadingsUSA.csv', index=False)
```
