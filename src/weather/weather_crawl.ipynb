```python
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
import pandas as pd



weather_data = []

for month in range(1,13):

    url = f'https://www.weatheri.co.kr/bygone/pastDB_tmp.php?start=2025-{month:02d}-01&orig_start=2025-1-01&end=2025-12-31&jijum_id=108'
    response = requests.get(url)
    response.encoding = 'utf-8'
    
    bs = BeautifulSoup(response.text, 'html.parser')
    
    weather_table = bs.select('table[width="614"]')
    rows = weather_table[0].select('tr')

    if month == 1:
        feature_names = [col.get_text(strip=True) for col in rows[0].select('td')]
    
    for row in rows[1:]:
        cols = row.select('td')
        row_info = [col.get_text(strip=True) for col in cols]

        weather_data.append(row_info)

    time.sleep(1.5)

weather_df = pd.DataFrame(weather_data, columns=feature_names)

weather_df.to_csv('./weather_charts_raw.csv',index=False,encoding='utf-8')
```
