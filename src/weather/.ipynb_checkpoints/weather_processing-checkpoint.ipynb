```python
import numpy as np
import pandas as pd

weather_df = pd.read_csv('./weather_charts.csv')

# 날짜 형태 바꾸기
weather_df['날짜'] = weather_df['날짜'].str.split(' ').str[0]
        # weather_df['날짜']: '날짜' 컬럼 선택
    	# .str: 해당 열의 개별 문자열에 접근
    	# .split(' '): 모든 행의 문자열을 공백 기준으로 자름
    	# .str[0] : 각 리스트의 0번 요소(날짜)만 추출
weather_df['날짜'] = pd.to_datetime(weather_df['날짜'], format='%Y년%m월%d일')

# 결측값 확인
# weather_df.info()

# 강수량, 신적설에 있는 '-' 값을 0.0으로 전환하기
print((weather_df == '-').sum())
weather_df.replace('-', 0.0, inplace=True)
weather_df = weather_df.astype({'강수량(mm)':'float64', '신적설(cm)':'float64'})
weather_df.info()

# 불쾌지수 값 추가
# 불쾌지수(DI): 9/5 * Ta - 0.55 (1-RH)* (9/5 *Ta - 26) + 32
    # Ta : 건구온도 (℃)
    # RH : 상대습도 (소수단위)
    # 계산할 때에는 건구 온도는 평균기온으로, 상대습도는 평균습도로 설정
weather_df['불쾌지수'] = round(1.8*weather_df['평균기온']-0.55*(1-weather_df['평균습도(%)']/100)*(1.8*weather_df['평균기온']-26)+32,2)
weather_df = weather_df[['날짜', '평균기온', '최고기온', '최저기온', '강수량(mm)', '신적설(cm)', '평균풍속(m/s)', '평균습도(%)', '운량(1/10)', '일조시간(Hr)', '불쾌지수', '날씨']]

weather_df.to_csv('../../data/processed_data/weather_charts_processed.csv',index=False,encoding='utf-8')
```
